## Basic Tensors and Operations

A tensor is a generalization of the scalar ($s \in \mathbb{R}$) the vector ($\vec{v} \in \mathbb{R}^n$) and the matrix ($M \in \mathbb{R}^{m \times n}$) to higher dimensions.

so a tensor is $t \in \mathbb{R}^{d_1 \times d_2 \times ... \times d_n}$ where $d_i$ is the dimension of the tensor in the $i$-th axis.

Note: That the tensor is not just a "real" number but can also be a complex number $t \in \mathbb{C}^{d_1 \times d_2 \times ... \times d_n}$ or even a boolean $t \in \{True, False\}^{d_1 \times d_2 \times ... \times d_n}$.

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
tf.config.list_logical_devices()
```

Creating a tensor with the identity matrix, the general identity matrix $I = \begin{bmatrix} 1 & \dots & 0 \\ \vdots & \ddots & \vdots \\ 0 & \dots & 1 \end{bmatrix}$ is a square matrix with ones on the diagonal and zeros elsewhere.

An important property of the identity matrix is that when it is multiplied by any other matrix, it leaves that matrix unchanged. For example, if $A$ is any matrix, then $I \cdot A = A \cdot I = A$.

```python
a = tf.eye(4)
print(a)
print(a.device)
```

Creating a tensor with all ones

```python
b = tf.ones((4, 4))
b
```

Tensor addition, multiplication, subtraction and matrix multiplication

```python
a + b
# or tf.add(a, b)
```

Note the difference between element-wise multiplication and matrix multiplication. Element-wise multiplication multiplies each corresponding element of the two tensors, while matrix multiplication involves multiplying the rows of the first tensor by the columns of the second tensor and summing the products to produce a new tensor.

```python
a * b
# or tf.multiply(a, b)
```

```python
a - b
# or tf.subtract(a, b)
```

```python
a @ b
# or tf.matmul(a, b)
```

```python
zeros = tf.zeros((3, 4))
zeros
```

Creating random tensor with uniform or normal distribution

```python
random_tensor = tf.random.uniform((3, 4), minval=0, maxval=10, dtype=tf.float32)
random_tensor
```

Testing the speed differences between GPU and CPU for matrix multiplication. Note: The speed difference will depend on the hardware you are running this on, and may not be significant for small matrices.

```python
%%timeit
# do a 1024x1024 matrix multiplication on the GPU time it

if tf.config.list_physical_devices('GPU'):
    with tf.device('GPU:0'):
        random_matrix1 = tf.random.uniform((1024, 1024), minval=0, maxval=10, dtype=tf.float32)
        random_matrix2 = tf.random.uniform((1024, 1024), minval=0, maxval=10, dtype=tf.float32)
        result = random_matrix1 @ random_matrix2
else:
    print("No GPU found. Running on CPU.")
```

```python
%%timeit
# do the same on the CPU and time it

with tf.device('CPU:0'):
    random_matrix1 = tf.random.uniform((1024, 1024), minval=0, maxval=10, dtype=tf.float32)
    random_matrix2 = tf.random.uniform((1024, 1024), minval=0, maxval=10, dtype=tf.float32)
    result = random_matrix1 @ random_matrix2
```

Here we see a matrix multiplication on GPU of two $1024\times 1024$ matrices took $770\mu s$ while the same operation on CPU took $3.86 ms$ which is about 5 times slower than the GPU. Note that the speed difference will depend on the hardware you are running this on, and may not be significant for small matrices.

Testing the speed differences between MPS and CPU for matrix multiplication in PyTorch. On macOS, PyTorch uses Apple's Metal Performance Shaders (MPS) backend instead of CUDA. The exact speedup will depend on your hardware, PyTorch version, and whether MPS is available.

```python
%%timeit
# do a 1024x1024 matrix multiplication on MPS and time it

if torch.backends.mps.is_available():
    random_matrix1 = torch.rand((1024, 1024), device='mps')
    random_matrix2 = torch.rand((1024, 1024), device='mps')
    result = random_matrix1 @ random_matrix2
    torch.mps.synchronize()
else:
    print("MPS not available. Skipping MPS benchmark.")
```

```python
%%timeit
# do the same on the CPU and time it

random_matrix1 = torch.rand((1024, 1024), device='cpu')
random_matrix2 = torch.rand((1024, 1024), device='cpu')
result = random_matrix1 @ random_matrix2
```

If MPS is available, you can compare its timing with the CPU timing above. Keep in mind that results may vary from run to run, and the first run on MPS can include initialization overhead. For accurate MPS timings, we explicitly synchronize after the matrix multiplication.
